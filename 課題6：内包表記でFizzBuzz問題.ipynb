```python
def fizzbuzz_check(i):
    # 3で割り切れて、なおかつ5で割り切れる数のとき
    if i % 3 == 0 and i % 5 == 0:
        return "FizzBuzz"
    # 3で割り切れる数のとき
    elif i % 3 == 0:
        return "Fizz"
    # 5で割り切れる数のとき 
    elif i % 5 == 0:
        return "Buzz"
    # それ以外の数値の場合は数値を文字列に変換
    else:
        return str(i)

# 1から100までの連続する整数に対するFizzBuzz判定を行う 
fizzbuzz_list = [fizzbuzz_check(i) for i in range(1,101)]

# FizzBuzz判定の結果をリストとして生成し、print()関数で全件を出力する
print(fizzbuzz_list)
```
