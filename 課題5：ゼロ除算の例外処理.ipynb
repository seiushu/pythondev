```python
try:
    input_a = input("割られる数を入力してください：")
    a = int(input_a)

    input_b = input("割る数を入力してください：")
    b = int(input_b)

    c = a / b

except ValueError:
    print(f"割られる数を入力してください：{input_a}")
    print("エラー: 数値を入力してください")
except ZeroDivisionError:
    print(f"割られる数を入力してください: {input_a}")
    print(f"割る数を入力してください：{input_b}")
    print("エラー: 0で割り算しないでください")
else:
    c = a / b
    print(f"割られる数を入力してください：{a}")
    print(f"割る数を入力してください：{b}")
    print(f"{a} ÷ {b} = {c}")

print("処理を終了します")
```
