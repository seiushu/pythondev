```python
score = int(input("税抜価格を入力してください："))
scorenew = round(score*1.10)

if scorenew >= 2000:
    print(f"税抜価格を入力してください：{score}")
    print("送料は無料です")    
    print(f"送料込みの価格は{scorenew}円です。")

else:
    print(f"税抜価格を入力してください：{score}")
    print("送料として350円かかります")  
    print(f"送料込みの価格は{scorenew + 350}円です。")
```
