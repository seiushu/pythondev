```python
count_score = 0
count_pepole = 0

while True:
    score_input = input("点数を入力してください:")
    score = int(score_input)
    print(f"点数を入力してください:{score}")

    if score == -1:
        break

    count_pepole += 1
    count_score += score

if count_pepole > 0:
    average_score = count_score / count_pepole

print(f"{count_pepole}人のテストの平均点は{average_score:.1f}点です。")
```
